```python
# ==========================================================
# Cognifyz Data Analysis Internship
# Level 3 - Task 2: Votes Analysis
# ==========================================================

# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ----------------------------------------------------------
# Load Dataset
# ----------------------------------------------------------

df = pd.read_csv(r"D:\2 Year iv sem\Python\Cognifyz-Data-Analysis\Dataset .csv")

print("="*60)
print("VOTES ANALYSIS")
print("="*60)

# ----------------------------------------------------------
# Basic Information
# ----------------------------------------------------------

print("\nTotal Restaurants:", len(df))

# ----------------------------------------------------------
# Highest Voted Restaurants
# ----------------------------------------------------------

highest_votes = df.sort_values(
    by="Votes",
    ascending=False
).head(10)

print("\nTop 10 Restaurants with Highest Votes:")
print(
    highest_votes[
        ["Restaurant Name",
         "City",
         "Votes",
         "Aggregate rating"]
    ]
)

# ----------------------------------------------------------
# Lowest Voted Restaurants
# ----------------------------------------------------------

lowest_votes = df.sort_values(
    by="Votes",
    ascending=True
).head(10)

print("\nTop 10 Restaurants with Lowest Votes:")
print(
    lowest_votes[
        ["Restaurant Name",
         "City",
         "Votes",
         "Aggregate rating"]
    ]
)

# ==========================================================
# Visualization 1
# Highest Voted Restaurants
# ==========================================================

plt.figure(figsize=(12,6))

sns.barplot(
    data=highest_votes,
    x="Restaurant Name",
    y="Votes",
    palette="viridis"
)

plt.title(
    "Top 10 Restaurants with Highest Votes",
    fontsize=15,
    fontweight="bold"
)

plt.xlabel("Restaurant")
plt.ylabel("Votes")

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

# ==========================================================
# Visualization 2
# Lowest Voted Restaurants
# ==========================================================

plt.figure(figsize=(12,6))

sns.barplot(
    data=lowest_votes,
    x="Restaurant Name",
    y="Votes",
    palette="magma"
)

plt.title(
    "Top 10 Restaurants with Lowest Votes",
    fontsize=15,
    fontweight="bold"
)

plt.xlabel("Restaurant")
plt.ylabel("Votes")

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

# ==========================================================
# Correlation Analysis
# ==========================================================

correlation = df["Votes"].corr(
    df["Aggregate rating"]
)

print("\nCorrelation Between Votes and Rating:")
print(round(correlation,3))

# ==========================================================
# Visualization 3
# Votes vs Rating
# ==========================================================

plt.figure(figsize=(10,6))

sns.scatterplot(
    data=df,
    x="Votes",
    y="Aggregate rating",
    hue="Aggregate rating",
    palette="coolwarm",
    s=80
)

plt.title(
    "Relationship Between Votes and Ratings",
    fontsize=15,
    fontweight="bold"
)

plt.xlabel("Votes")
plt.ylabel("Aggregate Rating")

plt.show()

# ==========================================================
# Top Rated & Most Voted Restaurants
# ==========================================================

top_voted = df.loc[df["Votes"].idxmax()]

print("\nRestaurant with Maximum Votes:")
print("Restaurant Name :", top_voted["Restaurant Name"])
print("City            :", top_voted["City"])
print("Votes           :", top_voted["Votes"])
print("Rating          :", top_voted["Aggregate rating"])

# ==========================================================
# Conclusion
# ==========================================================

print("\n" + "="*60)
print("CONCLUSION")
print("="*60)

print(
    f"\n1. Restaurant with highest votes: "
    f"{top_voted['Restaurant Name']}"
)

print(
    f"\n2. Maximum votes received: "
    f"{top_voted['Votes']}"
)

print(
    f"\n3. Correlation between votes and rating: "
    f"{round(correlation,3)}"
)

if correlation > 0:
    print(
        "\n4. Positive correlation found."
        "\n   Restaurants with more votes tend to have higher ratings."
    )

elif correlation < 0:
    print(
        "\n4. Negative correlation found."
        "\n   Restaurants with more votes tend to have lower ratings."
    )

else:
    print(
        "\n4. No significant correlation found."
    )

print(
    "\n5. Popular restaurants generally receive more customer engagement."
)

print(
    "\n6. High votes indicate popularity, while ratings indicate customer satisfaction."
)
```
