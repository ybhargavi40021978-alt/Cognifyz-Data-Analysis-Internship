```python
# ==========================================================
# Cognifyz Data Analysis Internship
# Level 2 - Task 1: Restaurant Ratings
# ==========================================================

# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

# ----------------------------------------------------------
# Load Dataset
# ----------------------------------------------------------

df = pd.read_csv(r"D:\2 Year iv sem\Python\Cognifyz-Data-Analysis\Dataset .csv")

print("Dataset Loaded Successfully!")
print(f"Total Restaurants: {len(df)}")

# ----------------------------------------------------------
# Distribution of Aggregate Ratings
# ----------------------------------------------------------

print("\nAggregate Rating Statistics:")
print(df["Aggregate rating"].describe())

# ----------------------------------------------------------
# Create Rating Ranges
# ----------------------------------------------------------

rating_bins = [0, 1, 2, 3, 4, 5]
rating_labels = [
    "0-1",
    "1-2",
    "2-3",
    "3-4",
    "4-5"
]

df["Rating Range"] = pd.cut(
    df["Aggregate rating"],
    bins=rating_bins,
    labels=rating_labels,
    include_lowest=True
)

rating_distribution = df["Rating Range"].value_counts().sort_index()

print("\nRating Distribution:")
print(rating_distribution)

# ----------------------------------------------------------
# Most Common Rating Range
# ----------------------------------------------------------

most_common_range = rating_distribution.idxmax()
most_common_count = rating_distribution.max()

print("\nMost Common Rating Range:")
print(f"{most_common_range} ({most_common_count} restaurants)")

# ----------------------------------------------------------
# Average Number of Votes
# ----------------------------------------------------------

average_votes = round(df["Votes"].mean(), 2)

print("\nAverage Votes Received:")
print(average_votes)

# ----------------------------------------------------------
# Result Table
# ----------------------------------------------------------

rating_result = pd.DataFrame({
    "Rating Range": rating_distribution.index,
    "Restaurant Count": rating_distribution.values
})

print("\nRating Distribution Table:")
print(rating_result)

# ----------------------------------------------------------
# Histogram of Aggregate Ratings
# ----------------------------------------------------------

plt.figure(figsize=(8,5))

plt.hist(
    df["Aggregate rating"],
    bins=10
)

plt.title("Distribution of Aggregate Ratings")
plt.xlabel("Aggregate Rating")
plt.ylabel("Number of Restaurants")

plt.tight_layout()
plt.show()

# ----------------------------------------------------------
# Bar Chart of Rating Ranges
# ----------------------------------------------------------

plt.figure(figsize=(8,5))

bars = plt.bar(
    rating_result["Rating Range"].astype(str),
    rating_result["Restaurant Count"]
)

plt.title("Restaurant Count by Rating Range")
plt.xlabel("Rating Range")
plt.ylabel("Number of Restaurants")

for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height,
        int(height),
        ha='center'
    )

plt.tight_layout()
plt.show()

# ----------------------------------------------------------
# Conclusion
# ----------------------------------------------------------

print("\nConclusion:")
print(
    f"The most common rating range is {most_common_range}, "
    f"with {most_common_count} restaurants."
)

print(
    f"Restaurants receive an average of "
    f"{average_votes} votes."
)

print(
    "The rating distribution helps identify how customer "
    "ratings are spread across restaurants."
)

# ----------------------------------------------------------
# Display Final Table
# ----------------------------------------------------------

rating_result

# The most common rating range is X–Y, indicating that most restaurants fall within this rating category. The average number of votes received by restaurants is Z. This suggests the general customer engagement level and overall rating distribution within the dataset.
```
