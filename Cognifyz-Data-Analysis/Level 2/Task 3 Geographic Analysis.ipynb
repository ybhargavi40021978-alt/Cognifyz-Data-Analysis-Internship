```python
# ==========================================================
# Cognifyz Data Analysis Internship
# Level 2 - Task 3: Geographic Analysis
# ==========================================================

# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ----------------------------------------------------------
# Load Dataset
# ----------------------------------------------------------

df = pd.read_csv(r"D:\2 Year iv sem\Python\Cognifyz-Data-Analysis\Dataset .csv")

print("="*60)
print("GEOGRAPHIC ANALYSIS OF RESTAURANTS")
print("="*60)

# ----------------------------------------------------------
# Check Latitude and Longitude
# ----------------------------------------------------------

geo_df = df.dropna(subset=['Latitude', 'Longitude'])

print("\nTotal Restaurants with Coordinates:", len(geo_df))

# ----------------------------------------------------------
# Basic Statistics
# ----------------------------------------------------------

print("\nLatitude Range:")
print("Min:", geo_df['Latitude'].min())
print("Max:", geo_df['Latitude'].max())

print("\nLongitude Range:")
print("Min:", geo_df['Longitude'].min())
print("Max:", geo_df['Longitude'].max())

# ==========================================================
# Visualization 1
# Restaurant Locations Scatter Plot
# ==========================================================

plt.figure(figsize=(12,8))

plt.scatter(
    geo_df['Longitude'],
    geo_df['Latitude'],
    alpha=0.6,
    c='blue',
    edgecolors='black'
)

plt.title("Restaurant Locations by Latitude and Longitude",
          fontsize=16,
          fontweight='bold')

plt.xlabel("Longitude")
plt.ylabel("Latitude")

plt.grid(True, linestyle='--', alpha=0.5)

plt.show()

# ==========================================================
# Visualization 2
# Density Plot (Restaurant Clusters)
# ==========================================================

plt.figure(figsize=(12,8))

sns.kdeplot(
    x=geo_df['Longitude'],
    y=geo_df['Latitude'],
    fill=True,
    cmap="Reds",
    thresh=0.05
)

plt.title("Restaurant Density Map (Cluster Analysis)",
          fontsize=16,
          fontweight='bold')

plt.xlabel("Longitude")
plt.ylabel("Latitude")

plt.show()

# ==========================================================
# Visualization 3
# Top Cities by Restaurant Count
# ==========================================================

top_cities = geo_df['City'].value_counts().head(10)

plt.figure(figsize=(12,6))

colors = sns.color_palette("viridis", len(top_cities))

bars = plt.bar(
    top_cities.index,
    top_cities.values,
    color=colors
)

plt.title("Top 10 Cities by Number of Restaurants",
          fontsize=16,
          fontweight='bold')

plt.xlabel("City")
plt.ylabel("Restaurant Count")

plt.xticks(rotation=45)

for bar in bars:
    height = bar.get_height()

    plt.text(
        bar.get_x() + bar.get_width()/2,
        height,
        str(height),
        ha='center',
        va='bottom'
    )

plt.tight_layout()
plt.show()

# ==========================================================
# Cluster Summary
# ==========================================================

print("\nTop 10 Cities with Highest Restaurant Count:")
print(top_cities)

most_dense_city = top_cities.idxmax()
most_dense_count = top_cities.max()

# ==========================================================
# Conclusion
# ==========================================================

print("\n" + "="*60)
print("CONCLUSION")
print("="*60)

print(f"\n1. Total restaurants analyzed: {len(geo_df)}")

print(f"\n2. Highest concentration of restaurants found in:")
print(f"   {most_dense_city}")

print(f"\n3. Number of restaurants in this city:")
print(f"   {most_dense_count}")

print("\n4. Scatter plot shows the geographical distribution of restaurants.")

print("\n5. Density map highlights areas where restaurants are clustered.")

print("\n6. Restaurants tend to concentrate in major urban locations rather than being evenly distributed.")

print("\n7. These clusters indicate areas with higher customer demand and commercial activity.")
```
