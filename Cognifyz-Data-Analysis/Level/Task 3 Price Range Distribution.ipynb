```python
# ==========================================================
# Cognifyz Data Analysis Internship
# Level 1 - Task 3: Price Range Distribution
# ==========================================================
#
# Task:
# 1. Visualize the distribution of price ranges among restaurants.
# 2. Calculate the percentage of restaurants in each price range.
#
# ==========================================================

# Import Required Libraries
import pandas as pd
import matplotlib.pyplot as plt

# ----------------------------------------------------------
# Load Dataset
# ----------------------------------------------------------

df = pd.read_csv(r"D:\2 Year iv sem\Python\Cognifyz-Data-Analysis\Dataset .csv")

print("Dataset Loaded Successfully!")
print(f"Total Restaurants: {len(df)}")

# ----------------------------------------------------------
# Count Restaurants in Each Price Range
# ----------------------------------------------------------

price_counts = df["Price range"].value_counts().sort_index()

print("\nRestaurant Count by Price Range:")
print(price_counts)

# ----------------------------------------------------------
# Calculate Percentage Distribution
# ----------------------------------------------------------

total_restaurants = len(df)

price_percentages = round((price_counts / total_restaurants) * 100, 2)

print("\nPercentage Distribution:")
print(price_percentages)

# ----------------------------------------------------------
# Create Result Table
# ----------------------------------------------------------

result = pd.DataFrame({
    "Price Range": price_counts.index,
    "Restaurant Count": price_counts.values,
    "Percentage (%)": price_percentages.values
})

print("\nResults:")
print(result)

# ----------------------------------------------------------
# Bar Chart Visualization
# ----------------------------------------------------------

plt.figure(figsize=(8, 5))

bars = plt.bar(
    result["Price Range"].astype(str),
    result["Restaurant Count"]
)

plt.title("Distribution of Restaurants by Price Range")
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")

for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height,
        int(height),
        ha='center',
        va='bottom'
    )

plt.tight_layout()
plt.show()

# ----------------------------------------------------------
# Pie Chart Visualization
# ----------------------------------------------------------

plt.figure(figsize=(7, 7))

plt.pie(
    result["Restaurant Count"],
    labels=result["Price Range"],
    autopct="%1.1f%%"
)

plt.title("Percentage Distribution of Price Ranges")
plt.show()

# ----------------------------------------------------------
# Conclusion
# ----------------------------------------------------------

most_common_range = result.loc[
    result["Restaurant Count"].idxmax(),
    "Price Range"
]

most_common_percentage = result.loc[
    result["Restaurant Count"].idxmax(),
    "Percentage (%)"
]

print("\nConclusion:")
print(
    f"Price Range {most_common_range} has the highest number of restaurants."
)
print(
    f"It represents {most_common_percentage}% of all restaurants."
)
print(
    "The distribution shows how restaurants are spread across "
    "different pricing categories."
)

# ----------------------------------------------------------
# Display Final Table
# ----------------------------------------------------------

result
```
