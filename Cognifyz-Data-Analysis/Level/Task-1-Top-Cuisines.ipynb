```python
# ==========================================================
# Cognifyz Data Analysis Internship
# Level 1 - Task 1: Top Cuisines
# ==========================================================
#
# Task:
# Determine the top three most common cuisines in the dataset.
# Calculate the percentage of restaurants that serve each of
# the top cuisines.
#
# ==========================================================

# Import Required Libraries
import pandas as pd
import matplotlib.pyplot as plt

# ----------------------------------------------------------
# Load Dataset
# ----------------------------------------------------------

df = pd.read_csv(r"D:\2 Year iv sem\Python\Cognifyz-Data-Analysis\Dataset .csv")

print("Dataset Loaded Successfully!")
print(f"Total Restaurants: {len(df)}")

# ----------------------------------------------------------
# Determine Top 3 Most Common Cuisines
# ----------------------------------------------------------

top_cuisines = df['Cuisines'].value_counts().head(3)

print("\nTop 3 Most Common Cuisines:")
print(top_cuisines)

# ----------------------------------------------------------
# Calculate Percentage of Restaurants
# ----------------------------------------------------------

total_restaurants = len(df)

cuisine_names = []
restaurant_counts = []
percentages = []

print("\nPercentage of Restaurants Serving Each Top Cuisine:\n")

for cuisine, count in top_cuisines.items():
    percentage = (count / total_restaurants) * 100

    cuisine_names.append(cuisine)
    restaurant_counts.append(count)
    percentages.append(round(percentage, 2))

    print(f"{cuisine}: {percentage:.2f}%")

# ----------------------------------------------------------
# Create Results Table
# ----------------------------------------------------------

result = pd.DataFrame({
    "Cuisine": cuisine_names,
    "Restaurant Count": restaurant_counts,
    "Percentage (%)": percentages
})

print("\nResults:")
print(result)

# ----------------------------------------------------------
# Bar Chart Visualization
# ----------------------------------------------------------

plt.figure(figsize=(8, 5))

bars = plt.bar(
    result["Cuisine"],
    result["Restaurant Count"]
)

plt.title("Top 3 Most Common Cuisines")
plt.xlabel("Cuisine")
plt.ylabel("Number of Restaurants")

for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height,
        int(height),
        ha='center',
        va='bottom'
    )

plt.xticks(rotation=15)
plt.tight_layout()
plt.show()

# ----------------------------------------------------------
# Pie Chart Visualization
# ----------------------------------------------------------

plt.figure(figsize=(7, 7))

plt.pie(
    result["Percentage (%)"],
    labels=result["Cuisine"],
    autopct="%1.1f%%"
)

plt.title("Percentage Distribution of Top 3 Cuisines")
plt.show()

# ----------------------------------------------------------
# Conclusion
# ----------------------------------------------------------

top_cuisine = result.iloc[0]["Cuisine"]
top_count = result.iloc[0]["Restaurant Count"]
top_percentage = result.iloc[0]["Percentage (%)"]

print("\nConclusion:")
print(f"{top_cuisine} is the most common cuisine in the dataset.")
print(f"It is served by {top_count} restaurants.")
print(f"It accounts for {top_percentage}% of all restaurants.")
print("The top three cuisines represent the most popular")
print("food preferences among restaurants in the dataset.")

# ----------------------------------------------------------
# Display Final Table
# ----------------------------------------------------------

result
```
