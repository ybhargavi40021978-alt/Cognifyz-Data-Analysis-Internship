```python
# ==========================================================
# Cognifyz Data Analysis Internship
# Level 1 - Task 2: City Analysis
# ==========================================================
#
# Task:
# 1. Identify the city with the highest number of restaurants.
# 2. Calculate the average rating for restaurants in each city.
# 3. Determine the city with the highest average rating.
#
# ==========================================================

# Import Required Libraries
import pandas as pd
import matplotlib.pyplot as plt

# ----------------------------------------------------------
# Load Dataset
# ----------------------------------------------------------

df = pd.read_csv(r"D:\2 Year iv sem\Python\Cognifyz-Data-Analysis\Dataset .csv")

print("Dataset Loaded Successfully!")
print(f"Total Restaurants: {len(df)}")

# ----------------------------------------------------------
# City with Highest Number of Restaurants
# ----------------------------------------------------------

city_counts = df["City"].value_counts()

top_city = city_counts.idxmax()
top_city_count = city_counts.max()

print("\nCity with Highest Number of Restaurants:")
print(f"{top_city} ({top_city_count} restaurants)")

# ----------------------------------------------------------
# Average Rating for Each City
# ----------------------------------------------------------

city_ratings = (
    df.groupby("City")["Aggregate rating"]
    .mean()
    .sort_values(ascending=False)
)

print("\nAverage Rating for Each City:")
print(city_ratings)

# ----------------------------------------------------------
# City with Highest Average Rating
# ----------------------------------------------------------

highest_rated_city = city_ratings.idxmax()
highest_rating = city_ratings.max()

print("\nCity with Highest Average Rating:")
print(f"{highest_rated_city} ({highest_rating:.2f})")

# ----------------------------------------------------------
# Create Results Table
# ----------------------------------------------------------

result = pd.DataFrame({
    "City": city_ratings.index,
    "Average Rating": city_ratings.values
})

print("\nTop 10 Cities by Average Rating:")
print(result.head(10))

# ----------------------------------------------------------
# Visualization 1:
# Top 10 Cities by Number of Restaurants
# ----------------------------------------------------------

top_10_restaurants = city_counts.head(10)

plt.figure(figsize=(10, 5))

bars = plt.bar(
    top_10_restaurants.index,
    top_10_restaurants.values
)

plt.title("Top 10 Cities by Number of Restaurants")
plt.xlabel("City")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=45)

for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height,
        int(height),
        ha='center',
        va='bottom'
    )

plt.tight_layout()
plt.show()

# ----------------------------------------------------------
# Visualization 2:
# Top 10 Cities by Average Rating
# ----------------------------------------------------------

top_10_ratings = city_ratings.head(10)

plt.figure(figsize=(10, 5))

bars = plt.bar(
    top_10_ratings.index,
    top_10_ratings.values
)

plt.title("Top 10 Cities by Average Restaurant Rating")
plt.xlabel("City")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)

for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height,
        f"{height:.2f}",
        ha='center',
        va='bottom'
    )

plt.tight_layout()
plt.show()

# ----------------------------------------------------------
# Conclusion
# ----------------------------------------------------------

print("\nConclusion:")
print(f"{top_city} has the highest number of restaurants ({top_city_count}).")
print(f"{highest_rated_city} has the highest average restaurant rating ({highest_rating:.2f}).")
print("City-wise analysis helps identify locations with")
print("high restaurant concentration and customer satisfaction.")

# ----------------------------------------------------------
# Display Final Table
# ----------------------------------------------------------

result.head(10)
```
